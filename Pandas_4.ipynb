## Sorting

sort ရဲ့ သဘောသဘာဝက character ဆိုရင် A to Z (Ascending) / Z to A (Descending)
မြန်မာစာဆိုရင်လည်း UTF-8 နဲ့ရေးထားတဲ့ font တွေဆိုရင် အက္ခရာစဉ်လို့ရပါတယ်။ မြန်မာစာမှာသင်ခဲ့ရတဲ့ အက္ခရာစဉ်နည်းအတိုင်းဖြစ်ပါတယ်။ ကိန်းဂဏန်းတွေဆိုရင် ကြီးစဉ်ငယ်လိုက်အတိုင်းပဲဖြစ်ပါတယ်။

```python
#ပထမဆုံးကျွန်တော်တို့ cut column မှာပါတဲ့ စာသားတွေကို sorting လုပ်ရအောင်ဗျာ။
#ကြည့်ရလွယ်အောင် df အသစ်တခုလုပ်ပြီး သူ့ထဲမှာ carat, cut, price, clarity ရယ်ကိုပဲ အဓိကထားလုပ်ရအောင်။
new_df=df[['carat','cut','clarity','price']]
new_df.head(4)
```

```python
new_df.sort_values('cut') #A2Z ပုံစံနဲ့ sort သွားပါတယ်။
```

```python
new_df.sort_values('cut',ascending=True)
#A2Z sorting လုပ်တာက ascending method ဖြစ်ပြီး ဒီမှာသုံးရင် ascending=True လို့ရေးပေးရပါတယ်။
```

```python
new_df.sort_values('cut',ascending=False)
# False ဆိုရင် Z2A descending ဖြစ်ပါတယ်။
# တခုသိထားရမှာက sort ထားတဲ့အတိုင်းသုံးချင်ရင် variable ထဲပြန်ပြီး assign လုပ်ဖို့လိုပါတယ်။
# ဆိုလိုတာက new_df=new_df.sort_values('cut',ascending=False)
```

```python
# ဂဏန်းတွေကို sort လုပ်ရအောင်
new_df.sort_values('price',ascending=False)#ဒါဆိုရင် price column မှာကြီးစဉ်ငယ်လိုက်ထွက်မှာဖြစ်ပါတယ်။
```
