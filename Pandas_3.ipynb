## Basic Filtering

ကျွန်တော်တို့ ပထမဆုံး clarity မှာ VS1 ဖြစ်တဲ့ data တွေကို စပြီးစစ်ထုတ်ပါ့မယ်။  (**ဒီအဆင့်မှာ မှာချင်တာက အမြဲတမ်း data types တွေကို စစ်ပြီးမှ filter လုပ်သင့်ပါတယ်။**) 

```python
df['clarity'].dtypes #ဒီလိုစစ်တဲ့အခါမှာ သူ့ရဲ့ datatype က object ဖြစ်သည်ကို တွေ့ရမှာဖြစ်ပါတယ်။ 
#အဲ့တာကြောင့် object ဆိုတာဟာ string ဖြစ်တဲ့အတွက် value filter လုပ်ရင် string ပုံစံနဲ့စစ်ပေးရမှာဖြစ်ပါတယ်။
```

```python
df['clarity']=='VS1'
#ဒီလိုဆိုရင် ကိုယ်ရချင်တဲ့ VS1 ရှိတဲ့နေရာတွေကို Boolean value(T/F)အဖြစ် တွေ့မြင်ရမှာဖြစ်ပါတယ်။
#True နေရာမှာဆိုရင် ကျနော်တို့ရချင်တဲ့ VS1 ရှိတယ်လို့သိထားရမှာဖြစ်ပြီး False က တခြားဟာတွေဖြစ်ပါတယ်။
```

VS1 ရဲ့ အရေအတွက်ကို သိချင်ရင်တော့ အပေါ်က ကျွန်တော်တို့လုပ်ထားတဲ့ function ကို value_counts() method ဖြင့် စစ်နိုင်ပါတယ်။

```python
(df['clarity']=='VS1').value_counts()
#ဒီနေရာမှာ True တန်ဖိုးက 8171 ဖြစ်သည့်အတွက် VS1 8171 ရှိတယ်ကိုတွေ့ရမှာဖြစ်ပါတယ်။
```

### value_counts() Method

value_counts() method ဟာ အသုံးဝင်တဲ့ method တခုဖြစ်ပါတယ်။ ကိုယ်သိချင်တဲ့ column ထဲက value တွေဘယ်လောက်ရှိလဲဆိုတာ သိချင်ရင် သုံးပါတယ်။

```python
df['clarity'].value_counts()
```

```python
# value_counts ကိုအသုံးပြုပြီး items အလိုက် ပါဝင်မှု ရာခိုင်နှုန်းကို တွက်လို့ရပါတယ်။
#dropna ကို false ပေးထားတာက values မရှိတဲ့ items တွေကို ဖြုတ်မပစ်အောင်ဖြစ်ပြီး normalizeက stats အသုံးအနှုန်းဖြစ်ပြီး ပါဝင်တဲ့အချိုးအစားတွက်ရင်သုံးပါတယ်။ 
#ပုံမှန်ဆိုရင် percent လုပ်ရင် 100 နဲ့စားရတယ်။ normalize လုပ်ထားတဲ့အဖြေကဟာကို percent ပြန်ပြောင်းမှာဆိုတော့ 100 နဲ့မြှောက်လိုက်တာဖြစ်ပါတယ်။

df.clarity.value_counts(normalize=True, dropna =False)*100
```

ကိုယ့်ရဲ့ filter လုပ်ထားတဲ့ တန်ဖိုးကို df[ ] (သို့) df.loc[ ] ထဲထည့်ပေးရင် Tabular ပုံစံနဲ့မြင်ရမှာဖြစ်ပါတယ်။

```python
df[df['clarity']=='VS1']
# df.loc[df['clarity']=='VS1']
```

filter စစ်ထားတဲ့ data တွေကို data frame အသစ်ထဲမှာ ပြန်ပြီး assign လုပ်လို့ရပါတယ်။ 

```python
new_df=df[df['clarity']=='VS1']
new_df
```

ကျွန်တော်တို့ရဲ့ Tabular output မှာ ဘယ်ဘက်က index number တွေ အစဉ်အလိုက်မဖြစ်တာကိုတွေ့ရမှာဖြစ်ပါတယ်။ အဲ့တာကြောင့်မို့ အစဉ်အလိုက်ဖြစ်အောင် ကျွန်တော်တို့အနေနဲ့ reset လုပ်ပေးဖို့လိုပါတယ်။ 

```python
new_df.reset_index()
```

Value တခုထဲကို filter လုပ်တာရပြီဆိုရင် value (၂)ခုကို filter လုပ်ကြည့်ရအောင်။ ကျွန်တော်တို့ ပထမနှစ်က သင်ခဲ့ရတဲ့ logic သဘောသဘာဝအတိုင်းပဲ value(၂)ခုစလုံးရချင်ရင် and နှင့်စစ်ရပြီး value(၂)ခုထဲက (၁)ခုပဲရချင်ရင် or နဲ့စစ်ရပါတယ်။

ဒီတော့ cut က value (၂) ခုကို ထုတ်ကြည့်ရအောင်။

```python
new_df[(new_df['cut']== 'Good') | (new_df['cut']=='Very Good')]
# | ဒီသင်္ကေတက or ရဲ့ သင်္ကေတဖြစ်ပါတယ်။
#cut column ထဲမှာ ကျွန်တော်တို့ရချင်တဲ့ value တွေဟာ တခုထက်ပိုတဲ့အတွက် or ကို သုံးရချင်းဖြစ်ပါတယ်။
```

```python
#filter လုပ်ရင်းနဲ့လည်း reset_index() ကို တိုက်ရိုက်သုံးလို့ရပါတယ်။
new_df[(new_df['cut']== 'Good') | (new_df['cut']=='Very Good')].reset_index()
```

ကျွန်တော်တို့အနေနဲ့ Good/ Very Good မဟုတ်တဲ့တန်ဖိုးတွေအားလုံးကို ရချင်ရင် filter ပြန်မလုပ်သင့်ဘဲ not ကို အသုံးပြုသင့်ပါတယ်။ not ရဲ့ သင်္ကေတက ~ ဖြစ်ပါတယ်။

```python
new_df[~((new_df['cut']== 'Good') | (new_df['cut']=='Very Good'))]
```

### မတူညီတဲ့ column (၂)ခုက value များကို Filter လုပ်ခြင်း

```python
#ဒီနေရာမှာကျွန်တော်က cut က Premium နဲ့ color က G ကို စစ်ထုတ်သွားမှာဖြစ်ပါတယ်။
df[(df['cut']=='Premium') & (df['color']=='G')]
```

```python
# တကယ်လို့ or ကို သုံးမယ်ဆိုရင် premium နဲ့ green ဖြစ်သမျှဟာတွေကို အားလုံးကို ထုတ်ပြမှာဖြစ်ပါတယ်။
df[(df['cut']=='Premium') | (df['color']=='G')]
```

### မတူညီတဲ့ column (၃)ခုက value များကို Filter လုပ်ခြင်း

Number တွေကို filter လုပ်ရင် equal(==),greater than(>), less than(<), greater than or equal (>=) လို အားလုံးနှင့်ရင်းနှီးတဲ့ဟာတွေပဲသုံးမှာပဲဖြစ်လို့ ထွေထွေထူးထူးမရှင်းတော့ပါဘူး။

```python
df[(df['cut']=='Premium') & (df['color']=='G')& (df['price']>=15000)]
```

Minimum/Maximum ဖြစ်သောနေရာများအား Filter လုပ်ခြင်း

```python
# ပထမဆုံးအနေနဲ့ column ရဲ့ max ကို အရင်ရှာရပါမယ်။
df['price'].max()
```

```python
# ကိုယ်ရှာထားတဲ့ value ကို မူလ column မှာပြန်ပြီး conditional စစ်လိုက်ရင်အဖြေက ရပါပြီ
df[df['price']==df['price'].max()]
```

```python
df[df['price']==df['price'].min()]
```
